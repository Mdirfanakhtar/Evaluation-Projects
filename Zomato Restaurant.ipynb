### Feature Engineering:

Rating Categories

```python
def categorize_rating(rating):
    if 4.5 <= rating <= 4.9:
        return 'Excellent'
    elif 4.0 <= rating <= 4.4:
        return 'Very Good'
    elif 3.5 <= rating <= 3.9:
        return 'Good'
    elif 2.5 <= rating <= 3.4:
        return 'Average'
    elif 2.0 <= rating <= 2.4:
        return 'Poor'
    else:
        return 'Not Categorized'  

# Apply the function to create a new 'Rating Category' column
df['Rating Category'] = df['Aggregate rating'].apply(categorize_rating)
```

Online Delivery Indicator

```python
# Convert 'Has Online delivery' to binary indicator
df['Online Delivery Indicator'] = (df['Has Online delivery'] == 'Yes').astype(int)
```

City Population Categories (Hypothetical Example)

```python
def categorize_city_population(population):
    if population > 1000000:
        return 'Metropolitan'
    elif 500000 <= population <= 1000000:
        return 'Urban'
    else:
        return 'Rural'
df['Country Votes Category'] = df['Votes'].apply(categorize_city_population)
```

Numeric Feature Scaling (if needed)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Average Cost for two', 'Votes']] = scaler.fit_transform(df[['Average Cost for two', 'Votes']])
```

Geographical Features

```python
# Define latitude and longitude of the reference location
latitude_ref = 40.7128  # Example latitude of New York City
longitude_ref = -74.0060  # Example longitude of New York City

from geopy.distance import geodesic

def calculate_distance(lat1, lon1, ref_location):
    return geodesic((lat1, lon1), ref_location).km
df['Distance from Reference'] = df.apply(lambda row: calculate_distance(row['Latitude'], row['Longitude'], (latitude_ref, longitude_ref)), axis=1)
```

### Outlier Detection:

Z-score Method:

```python
from scipy.stats import zscore

z_scores = zscore(df[['Average Cost for two', 'Votes']])
threshold = 3
outliers = (np.abs(z_scores) > threshold).any(axis=1)
df_no_outliers = df[~outliers]
```

Interquartile Range (IQR) Method:

```python
# Compute the first quartile (Q1) and third quartile (Q3)
Q1 = df['Votes'].quantile(0.25)
Q3 = df['Votes'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (df['Votes'] < lower_bound) | (df['Votes'] > upper_bound)

df_no_outliers = df[~outliers]
```

### Dimensionality Reduction

PCA for dimensionality reduction

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features = ['Average Cost for two', 'Votes']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
n_components = 2  
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(scaled_features)
columns = [f'PC{i+1}' for i in range(n_components)]
df_pca = pd.DataFrame(data=principal_components, columns=columns)
df_pca_final = pd.concat([df.drop(columns=features), df_pca], axis=1)
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

### Handling Imbalanced Data:

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE

numerical_cols = ['Average Cost for two', 'Votes']  
categorical_cols = ['City', 'Cuisines']  

X = df[numerical_cols + categorical_cols]
y = df['Rating Category']
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_cols),  
        ('cat', OneHotEncoder(), categorical_cols)  
    ],
    remainder='drop'  
)

X_encoded = preprocessor.fit_transform(X)
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_encoded, y)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)
print(classification_report(y_test, y_pred))
```
